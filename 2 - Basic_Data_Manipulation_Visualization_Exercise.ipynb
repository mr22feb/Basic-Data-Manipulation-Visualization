### 50. Write a Python program to convert a string to datetime.
Sample String : Jul 1 2014 2:43PM 

Expected Output : 2014-07-01 14:43:00

```python
sample_str = 'Jul 1 2014 2:43PM'

formatted_datetime = dt.datetime.strptime(sample_str, '%b %d %Y %I:%M%p')
output_string = formatted_datetime.strftime('%Y-%m-%d %H:%M:%S')

print(output_string)
```

### 52. Write a Python program to convert unix timestamp string to readable date.

### 54. Write a Python program to get days between two dates.

```python
data_2019 = company_data[company_data['Month'].str.contains('2019')]

# Pivot the data to calculate total sales (quantity) for each product in 2019
pivot_table_2019 = data_2019.pivot_table(index='Month', values=['FaceCream', 'FaceWash', 'ToothPaste', 'Soap', 'Shampo', 'Moisturizer'], aggfunc='sum')

# Calculate total sales for each product in 2019
total_sales_2019 = pivot_table_2019.sum()

# Create a Pie chart for total sales data in 2019
plt.figure(figsize=(8, 8))
plt.pie(total_sales_2019, labels=total_sales_2019.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6'])
plt.title('Total Sales Data for 2019')
plt.show()
```
